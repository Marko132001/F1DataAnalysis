```python
import fastf1
from matplotlib import pyplot as plt
```

```python
session = fastf1.get_session(2024, 'Sao-Paolo', 'Race')
session.load()
```

```python
session.laps.pick_drivers("VER")
print(session.get_driver('1'))
print(session.drivers)
```

```python
session.laps.get_weather_data()
```

```python
verstapen_data = session.laps.pick_driver('VER').pick_fastest().get_car_data()

t = verstapen_data["Time"]
vCar = verstapen_data["Speed"]

print(t)
fig, ax = plt.subplots()
ax.plot(t, vCar, label='Fast')
ax.set_xlabel('Time')
ax.set_ylabel('Speed [Km/h]')
ax.legend()
plt.show()
```
